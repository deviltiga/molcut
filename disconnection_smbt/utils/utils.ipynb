```python
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

# 分析 NPY 文件
def analyze_npy_files(directory):
    size_distribution = defaultdict(int)
    
    for filename in os.listdir(directory):
        if filename.endswith('.npy'):
            filepath = os.path.join(directory, filename)
            data = np.load(filepath)
            if len(data.shape) == 2 and data.shape[0] == data.shape[1]:
                token_length = data.shape[0]
                size_distribution[token_length] += 1
    
    return size_distribution

# 设置目录路径
directory = "../data/final_data/input_matrix"

# 分析文件
distribution = analyze_npy_files(directory)

# 打印基本统计信息
print("Token长度分布：")
for token_length, count in sorted(distribution.items()):
    print(f"Token长度 {token_length}: {count} 个文件")

total_files = sum(distribution.values())
print(f"\n总文件数：{total_files}")
print(f"最小Token长度：{min(distribution.keys())}")
print(f"最大Token长度：{max(distribution.keys())}")
print(f"平均Token长度：{sum(k*v for k,v in distribution.items()) / total_files:.2f}")

# 绘制柱状图
token_lengths = list(distribution.keys())
counts = list(distribution.values())

plt.figure(figsize=(12, 6))
plt.bar(token_lengths, counts)
plt.xlabel('Token 长度')
plt.ylabel('文件数量')
plt.title('NPY文件 Token 长度分布')
plt.xticks(rotation=45)
plt.tight_layout()

# 显示图表
plt.show()
```

```python
import os
import numpy as np

# 设置目录路径
directory = '../data/final_data/emb/'

# 处理文件
def process_npy_files(dir_path):
    for filename in os.listdir(dir_path):
        if filename.endswith('.npy'):
            file_path = os.path.join(dir_path, filename)
            
            # 加载NPY文件
            data = np.load(file_path)
            
            # 检查形状
            if data.shape[:3] == (8, 1, 8):
                # 使用squeeze()去除大小为1的维度
                data_processed = np.squeeze(data, axis=1)
                
                # 保存处理后的文件（覆盖原文件）
                np.save(file_path, data_processed)
                
                print(f"处理文件: {filename}")
                print(f"  原始形状: {data.shape}")
                print(f"  处理后形状: {data_processed.shape}")
            else:
                print(f"警告: {filename} 的形状不符合预期 {data.shape}")

# 运行处理
process_npy_files(directory)

print("处理完成。")

# 验证结果
print("\n验证结果:")
for filename in os.listdir(directory):
    if filename.endswith('.npy'):
        data = np.load(os.path.join(directory, filename))
        print(f"{filename}: {data.shape}")
    break
```
